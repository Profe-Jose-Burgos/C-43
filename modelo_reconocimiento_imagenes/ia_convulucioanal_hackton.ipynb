```python
import shutil
import os
carpeta_fuente = '/content/deportivo'
carpeta_destino = '/content/dataset/deportivos'

def cambio_de_carpeta(carpeta_fuente,carpeta_destino):
  imagenes = os.listdir(carpeta_fuente)

  for i, nombreimg in enumerate(imagenes):
    if i < 474:
      shutil.copy(carpeta_fuente + '/' + nombreimg, carpeta_destino + '/' + nombreimg)
```

```python
carpeta_fuente = '/content/camioneta'
carpeta_destino = '/content/dataset/camionetas'
cambio_de_carpeta(carpeta_fuente,carpeta_destino)

carpeta_fuente = '/content/coupe'
carpeta_destino = '/content/dataset/coupe'
cambio_de_carpeta(carpeta_fuente,carpeta_destino)

carpeta_fuente = '/content/pickup'
carpeta_destino = '/content/dataset/pickups'
cambio_de_carpeta(carpeta_fuente,carpeta_destino)

carpeta_fuente = '/content/sedan'
carpeta_destino = '/content/dataset/sedanes'
cambio_de_carpeta(carpeta_fuente,carpeta_destino)
```

```python

from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np

data_gen = ImageDataGenerator(
    rescale=1. / 255,
    rotation_range = 30,
    width_shift_range = 0.25,
    height_shift_range = 0.25,
    shear_range = 15,
    zoom_range = [0.5, 1.5],
    validation_split=0.2
    
)
```

```python
x_train = data_gen.flow_from_directory('/content/dataset', target_size = (224,224),
                                                     batch_size = 32, shuffle = True, subset = 'training')

x_test = data_gen.flow_from_directory('/content/dataset', target_size = (224,224),
                                                     batch_size = 32, shuffle = True, subset = 'validation')
```

```python
import tensorflow as tf
import tensorflow_hub as hub
url = "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4"
mobilnet = hub.KerasLayer(url, input_shape = (224,224,3))

```

```python
mobilnet.trainable = False
```

```python
model_hackton = tf.keras.Sequential([mobilnet,tf.keras.layers.Dense(5, activation = 'softmax')
                                     ])
```

```python
model_hackton.compile(
    optimizer = 'adam',
    loss = 'categorical_crossentropy',
    metrics = ['accuracy']  

)
```

```python
epoch = 50

hsitoria = model_hackton.fit(x_train, epochs = epoch,batch_size= 32,
                             validation_data = x_test,
                  )
```

```python
model_hackton.save("IA_hackton.h5",hsitoria)
print("modelo creado")
```
